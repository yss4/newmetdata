### Wendy's data merge

```python
wendy_df_1 = pd.read_csv('./data/voter_registration_county_2015_cleaned.csv')
for col in wendy_df_1.columns[1:]:
    wendy_df_1[col] = wendy_df_1[col].str.replace(',', '')
    wendy_df_1[col] = wendy_df_1[col].astype(int)

wendy_df_1
```

```python
wendy_df_2 = pd.read_csv('./data/unemployment.csv')
wendy_df_2.pop('Unnamed: 0')
area = wendy_df_2.pop('Area')
wendy_df_2.insert(0, "County", area)
wendy_df_2['County'] = wendy_df_2['County'].str.replace(' County', '')

for col in wendy_df_2.columns:
#     print col
    if wendy_df_2[col].dtype == np.object:
        if 'Employment' in col or 'Unemployment' in col or 'Unemployment Rate' in col:
            wendy_df_2[col] = wendy_df_2[col].str.replace(',', '')
            wendy_df_2[col] = wendy_df_2[col].astype(float)

wendy_df_2
```

```python
join_df_with_wendy = pd.merge(wendy_df_1, wendy_df_2, on='County', how='left')
print cl.Counter(pd.isnull(wendy_df_2['Unemployment Rate_y']))
join_df_with_wendy
```
